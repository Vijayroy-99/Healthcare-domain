Where is the leakage concentrated?

```python
%sql
SELECT hcpcs_code, line_count, total_allowed, total_paid,
       total_variance, payment_rate, zero_pay_pct
FROM claims.gold.agg_procedure_payment
ORDER BY total_variance DESC
LIMIT 20;
```

Which procedures are most often zero-paid?

```python
%sql
SELECT hcpcs_code, line_count, zero_pay_pct, total_allowed
FROM claims.gold.agg_procedure_payment
WHERE zero_pay_pct > 20 AND line_count > 500
ORDER BY total_allowed DESC;
```

How much does the patient carry?

```python
%sql
SELECT
  CASE WHEN patient_responsibility = 0 THEN 'None'
       WHEN patient_responsibility < 50 THEN 'Under $50'
       WHEN patient_responsibility < 200 THEN '$50-$200'
       ELSE 'Over $200' END AS burden_band,
  COUNT(*) AS lines,
  ROUND(AVG(line_allowed), 2) AS avg_allowed
FROM claims.silver.service_line
GROUP BY 1;
```
